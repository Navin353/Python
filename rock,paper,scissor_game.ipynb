```python
#importing tkinter and random

from tkinter import *
import random
```

```python
#initilize window

c= Tk()
c.geometry('400x400')
c.title('Rock,Paper,Scissors')
c.config(bg ='seashell3')

#labeling

Label(c, text = 'Rock, Paper ,Scissors' , font='arial 20 bold', bg = 'seashell2').pack()

#For user choice

user_take = StringVar()
Label(c, text = 'choose any one: rock:r, paper:p ,scissors:s' , font='arial 13 bold', bg = 'seashell2').place(x = 20,y=70)
Entry(c, font = 'arial 15', textvariable = user_take , bg = 'antiquewhite2').place(x=90 , y = 130)

#computer pick

comp_pick = random.randint(1,3)
if comp_pick == 1:
    comp_pick = 'r'
elif comp_pick ==2:
    comp_pick = 'p'
else:
    comp_pick = 's'

#Game    
Result = StringVar()
def play():
    user_pick = user_take.get()
    if user_pick == comp_pick:
        Result.set('tie,you both select same')
    elif user_pick == 'r' and comp_pick == 'p':
        Result.set('you loose,computer select paper')
    elif user_pick == 'r' and comp_pick == 's':
        Result.set('you win,computer select scissors')
    elif user_pick == 'p' and comp_pick == 's':
        Result.set('you loose,computer select scissors')
    elif user_pick == 'p' and comp_pick == 'r':
        Result.set('you win,computer select rock')
    elif user_pick == 's' and comp_pick == 'r':
        Result.set('you loose,computer select rock')
    elif user_pick == 's' and comp_pick == 'p':
        Result.set('you win ,computer select paper')
    else:
        Result.set('invalid: choose any one -- rock:r, paper:p, scissors:s')  

#reset function
def Reset():
    Result.set("") 
    user_take.set("")

#exit    
def Exit():
    c.destroy()

#Buttons
Entry(c, font = 'arial 10 bold', textvariable = Result, bg ='antiquewhite2',width = 50,).place(x=25, y = 250)
Button(c, font = 'arial 13 bold', text = 'PLAY'  ,padx =5,bg ='seashell4' ,command = play).place(x=150,y=190)
Button(c, font = 'arial 13 bold', text = 'RESET'  ,padx =5,bg ='seashell4' ,command = Reset).place(x=70,y=310)
Button(c, font = 'arial 13 bold', text = 'EXIT'  ,padx =5,bg ='seashell4' ,command = Exit).place(x=230,y=310)


c.mainloop()    
```
