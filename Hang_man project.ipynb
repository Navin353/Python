# HANGMAN GAME

```python
#importing random
import random

def word():
    l=["wikipedia","pneumonia","encylopedia","manifestation","pandemic","harbinger"]
    ch=random.choice(l)
    return ch.upper()

def blanks(w):
    return("_ "*len(w))



def hangman():
    h=[""" 
    ------
    |     |
    |     O
    |
    |
    |
    |
    --------
    """,
    """
    ------
    |     |
    |     O
    |     |
    |     
    |
    |
    --------
    """,
    """
    ------
    |     |
    |     O
    |     |
    |     |
    |
    |
    --------
    """,
    
    """
    ------
    |     |
    |     O
    |    \|
    |     |
    |      
    |
    --------
    """,
    
    """
    ------
    |     |
    |     O
    |    \|/
    |     |
    |      
    |
    --------
    """,
    """
    ------
    |     |
    |     O
    |    \|/
    |     |
    |    /
    |
    --------
    """,
    """
    ------
    |     |
    |     O
    |    \|/
    |     |
    |    / \ |
    |
    --------
    """
   ]
    return h
        
        
        
        
        
        
print("Welcome to HANGMAN GAME")
print("WORD TO BE GUESSED")
w=word()
b=blanks(w)
print(b)
l_blank=b.split(" ")

ha=hangman()
t=0
s=""
while True:
    n=input("Guess a letter?")
   
    if n.upper() in list(s):
        print("choose another letter,already choosen")
        continue
    else:
        if n.upper() in list(w):
            print("correct guess")
            index=[i for i,l in enumerate(w) if l==n.upper()]
        
            for i in index:
                l_blank[i]=n.upper()
                s=" ".join(l_blank)
            print(s)
        else:
            print(ha[t])
            if(t<7):
                print(" you guessed it wrong ,please try again!!")
                t+=1
            if(t==7):
                print("Man Hanged! GAME OVER")
                break
        
        s1="".join(l_blank)
        if(s1==w):
            print("No. Of tries done:",t)
            break
```
